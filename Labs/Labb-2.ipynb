```python
import math 
import matplotlib.pyplot as plt 

filename = ("../Labs/datapoint.txt")
data = []
with open (filename, "r") as file:
    next(file) 
    for line in file.readlines(): 
     width, height, label = line.strip().split(", ")
     data.append ((float(width), float(height), int(label)))

test_points= [
 (25, 32),
 (24.2, 31.5),
 (22, 34),
 (20.5, 34)]

print ("klaccifiering av testpunkter:")
for test_point in test_points:
    # Hitta n√§rmaste granne
    min_distance = float('inf')
    nearest_neighbor = None
    for point in data:
        distance = math.sqrt((test_point[0] - point[0])**2 + (test_point[1] - point[1])**2)
        if distance < min_distance:
            min_distance = distance
            nearest_neighbor = point

    result = "Pikachu" if nearest_neighbor[2] == 1 else "Pichu"
    print(f"Sample with (width, height): {test_point} classified as {result}")

pichu_data = [(x[0], x[1]) for x in data if x[2] == 0]
pikachu_data = [(x[0], x[1]) for x in data if x[2] == 1]

plt.figure(figsize=(10, 6))
plt.scatter(*zip(*pichu_data), color='blue', label='Pichu')
plt.scatter(*zip(*pikachu_data), color='red', label='Pikachu')
plt.scatter(*zip(*test_points), color='green', marker='x', s=100, label='Test Points')
plt.xlabel('Width (cm)')
plt.ylabel('Height (cm)')
plt.title('Pichu vs Pikachu Classification')
plt.legend()
plt.grid(True)
plt.show()
   


```
